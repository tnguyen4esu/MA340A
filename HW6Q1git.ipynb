```python
n = int(input("Enter number wanted to Collatz Conjecture: "))

G = DiGraph()


for s in range (1,n+1):
    G.add_vertex(s)
    #print(s)

while n > 0:
    i = n
    while i > 1:
        if i%2 == 0:
            newEdge = i//2
            G.add_edge((i, newEdge))
            i = i//2
        else:
            newEdge = 3*i + 1
            G.add_edge((i, newEdge))
            i = 3*i+1
    n = n-1

G.show()
```
